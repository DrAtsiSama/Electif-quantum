```python
from qiskit import *
```

```python
# Creer circuit quantique
```

```python
circ = QuantumCircuit(3)
```

```python
# Afficher le circuit
```

```python
circ.draw()
```

```python
circ.draw('mpl')
```

```python
# Intriquons les qubits 1 et 2
```

```python
circ.cx(1,2)
```

```python
circ.draw('mpl')
```

```python
# utilisons un simulateur quantique
```

```python
# importe AER = module simulator
```

```python
from qiskit import Aer
```

```python
# On créer notre moteur de calcul
```

```python
backend = Aer.get_backend('statevector_simulator')
```

```python
# On lance le calcul quantique
```

```python
job = backend.run(circ)
```

```python
# Affichons le resultat
```

```python
result = job.result()
```

```python
# Vecteurs d'etat
```

```python
outputstate = result.get_statevector(circ, decimals=3)
```

```python
print(outputstate)
```

```python
# afficher histogramme
```

```python
from qiskit.visualization import plot_state_city
```

```python
plot_state_city(outputstate)
```
